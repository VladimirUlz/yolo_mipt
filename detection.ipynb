```python
from ultralytics import YOLO
import cv2
```

```python
# Загрузка предобученной модели YOLOv8
model = YOLO('yolo11n.pt')  # YOLOv8 Nano
```

```python
import matplotlib.pyplot as plt

# Загружаем изображение
image_path = "bathroom.jpg"  # Укажите путь к вашему изображению
results = model(image_path)  # Прогоняем изображение через модель

# Обрабатываем каждый результат
for result in results:
    # Сохраняем изображение с аннотациями
    annotated_image = result.plot()  # Рисуем аннотации на изображении
    cv2.imwrite("annotated_bathroom.jpg", annotated_image)  # Сохраняем в файл

    # Отображаем изображение с использованием matplotlib
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB))  # Конвертируем BGR -> RGB
    plt.axis('off')  # Убираем оси для красоты
    plt.show()  # Показываем изображение
```
